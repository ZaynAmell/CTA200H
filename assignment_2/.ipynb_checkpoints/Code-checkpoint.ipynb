```python
import math
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import solve_ivp
```

Question 1

```python
def deriv_1(f,x0,h): #The derivative for the first method
    return (f(x0+h)-f(x0))/h
def deriv_2(f,x0,h): #The derivative for the second method
    return (f(x0+h)-f(x0-h))/(2*h)
```

```python
#Create four list to store different values
d1=[]
d2=[]
e1=[]
e2=[]
index=[]

#Loop for results with different h
f=math.sin
a=math.cos(0.1)
for h in np.arange(0.01,1.00,0.01):
    index.append(h/100)
    d1.append(deriv_1(lambda f:math.sin(f),1,h))
    d2.append(deriv_2(lambda f:math.sin(f),1,h))
    e1.append(abs(deriv_1(lambda f:math.sin(f),1,h)-a)/a)
    e2.append(abs(deriv_2(lambda f:math.sin(f),1,h)-a)/a)
    
#Draw the corresponding plots
plt.loglog(index,e1)
plt.loglog(index,e2)
plt.savefig('Question_1.png')
plt.show()
```

From the above plot, the error for the second method is significantly smaller than the first method.
The slope from the above plot represents the error increasing speed.

Question 2

```python
#assign all coordinates to different lists
a_x=[] #real part of diverging points
a_y=[] #imaginary part of diverging points
a_i=[] #the number of iteration to reach NaN
b_x=[] #real part of bounded points
b_y=[] #imaginary part of bounded points

for x in np.linspace(-2,2,100): #Creating real parts
    for y in np.linspace(-2,2,100): #Creating imaginary parts
        #initial conditions
        c=x+y*1j
        z=0
        #starting iteration
        for i in range(100):
            z=z**2+c
            #when the points is diverged, stop the loop and record data
            if (math.isnan(z.real)|math.isnan(z.imag)):
                a_x.append(x)
                a_y.append(y)
                a_i.append(i)
                break
        #for bounded points, record the data
        else:
            b_x.append(x)
            b_y.append(y)
#plot all the graphs
fig,ax=plt.subplots(1,2,dpi=100,figsize=(10,4))
ax[0].scatter(a_x,a_y)
ax[0].scatter(b_x,b_y)
ax[0].set_xlabel("real")
ax[0].set_ylabel("imaginary")
ax[0].set_title("Divergent vs bounded points")

ax[1].scatter(a_x,a_y,c=a_i)
ax[1].set_xlabel("real")
ax[1].set_ylabel("imaginary")
fig.colorbar(ax[1].scatter(a_x,a_y,c=a_i))
ax[1].set_title("Number of iterations to reach diverged")
plt.savefig("Question_2.png")
plt.show()
```

Question 3

```python
#Define SIR model
def SIR(t,y,B,Y,N):
    S, I, R =y
    return [-B*S*I/N, B*S*I/N-Y*I, Y*I]

#Define a solver to take different parameters
def solve(N,B,Y):
    return solve_ivp(lambda t,y:SIR(t,y,B,Y,N),t_span=[0,200],y0=[999,1,0],t_eval=np.linspace(0,200,N))

solve_1=solve(1000,0.5,0.2)
solve_2=solve(1000,0.4,0.2)
solve_3=solve(1000,0.4,0.3)
solve_4=solve(1000,0.5,0.3)

#Plot all four solution in separate figures
fig,ax=plt.subplots(2,2,dpi=100,figsize=(10, 7))

ax[0,0].plot(solve_1.t, solve_1.y[0],'r',solve_1.t, solve_1.y[1],'b',solve_1.t, solve_1.y[2],'g')
ax[0,0].set_xlabel("time")
ax[0,0].set_ylabel("Population")
ax[0,0].legend(labels=["S","I","R"])

ax[1,0].plot(solve_2.t, solve_2.y[0],'r',solve_2.t, solve_2.y[1],'b',solve_2.t, solve_2.y[2],'g')
ax[1,0].set_xlabel("time")
ax[1,0].set_ylabel("Population")
ax[1,0].legend(labels=["S","I","R"])

ax[0,1].plot(solve_3.t, solve_3.y[0],'r',solve_3.t, solve_3.y[1],'b',solve_3.t, solve_3.y[2],'g')
ax[0,1].set_xlabel("time")
ax[0,1].set_ylabel("Population")
ax[0,1].legend(labels=["S","I","R"])

ax[1,1].plot(solve_4.t, solve_4.y[0],'r',solve_4.t, solve_4.y[1],'b',solve_4.t, solve_4.y[2],'g')
ax[1,1].set_xlabel("time")
ax[1,1].set_ylabel("Population")
ax[1,1].legend(labels=["S","I","R"])

plt.savefig("Question_3(a).png")
plt.show()
```

Bouns
<br>
<br>For death, there should be a new parameter $\alpha$ to determine to death rate. Since the death can only come from the infected, the differential equation for death rate should be as follow:
<br><div align="center">$\large\frac{dD}{dt} =\alpha I$<br>
<div align="left">Now, because the total population is set to remain unchanged meaning $\frac{dN}{dt}=0$ and the increased infected should deduct the death rate. Therefore the infected equation should be modified as follow:
<br><center>$\large\frac{dI}{dt}=-\frac{\beta SI}{N} -\gamma I -\alpha I$

```python
#New function or SIRD
def SIRD(t,y,B,Y,N,A):
    S, I, R, D =y
    return [-B*S*I/N, B*S*I/N-Y*I-A*I, Y*I, A*I]
#new solver
def new_solve(N,B,Y,A):
    return solve_ivp(lambda t,y:SIRD(t,y,B,Y,N,A),t_span=[0,200],y0=[999,1,0,0],t_eval=np.linspace(0,200,N))

new_solver=new_solve(1000,0.5,0.2,0.1)

#plot new model
fig,ax=plt.subplots(1,2,dpi=100,figsize=(10,4))
ax[0].plot(new_solver.t, new_solver.y[0], 'r', new_solver.t, new_solver.y[1],'b',
         new_solver.t, new_solver.y[2], 'g', new_solver.t, new_solver.y[3],'m')
ax[0].set_xlabel("time")
ax[0].set_ylabel("Population")
ax[0].legend(labels=["S","I","R","D"])
ax[0].set_title("SIRD model")

ax[1].plot(solve_1.t, solve_1.y[0],'r',solve_1.t, solve_1.y[1],'b',solve_1.t, solve_1.y[2],'g')
ax[1].set_xlabel("time")
ax[1].set_ylabel("Population")
ax[1].legend(labels=["S","I","R"])
ax[1].set_title("SIR model")

plt.savefig("Question_3(b).png")
plt.show()
```

This new equation has the same parameter with the first set from SIR model, plus a death rate parameter $\alpha=0.1$. From the figures above, the time it takes to re-establish stability does not change. However, due to the death factor, the maximum infected population is lower. The remaining susceptible population is much larger than SIR model, because it took fewer people to reach herd immunity.
